# Part 1: Analyzing Best-Seller Trends Across Product Categories
Objective: Understand the relationship between product categories and their best-seller status.

## Crosstab Analysis:

Create a crosstab between the product category and the isBestSeller status.

Are there categories where being a best-seller is more prevalent?

Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.

```python
# Generating a crosstab for 'category' and 'isBestSeller'
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])

crosstab_result
```

```python
crosstab = pd.crosstab(df['category'], df['isBestSeller'], margins=True, normalize='index')

# Only consider cases where 'isBestSeller' equals True
best_seller_proportions = crosstab[True]

# Sort categories based on the proportion of best-sellers
sorted_categories = best_seller_proportions.sort_values(ascending=False)
sorted_categories.head(15) # Grocery, Smart Home, Health and Mobile Phone accesories are the categories where being a best-seller is more prevalent
```

## Statistical Tests:

Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.

Compute Cramér's V to understand the strength of association between best-seller status and category.

```python
from scipy.stats import chi2_contingency

# Perform the Chi-square test
chi2_statistic, chi2_p_value, dof, expected_freq = chi2_contingency(crosstab_result)

chi2_statistic , chi2_p_value
```

```python
# Compute Cramér's V
from scipy.stats.contingency import association

# Computing the association between variables in 'crosstab_result' using the "cramer" method
association(crosstab_result, method="cramer")
```

# Part 2: Exploring Product Prices and Ratings Across Categories and Brands
Objective: Investigate how different product categories influence product prices.

## Preliminary Step: Remove outliers in product prices.

For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. The next steps will be done with the dataframe without outliers.

Hint: you can check the last Check For Understanding at the end of the lesson EDA Bivariate Analysis for a hint on how to do this.


```python
# From lesson: Here's a Python function that implements Tukey's test for outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

Q1, Q3, IQR
```

```python
# Define bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
```

```python
# Filter the DataFrame to remove outliers
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# Display summary of the cleaned DataFrame
print(f"Original DataFrame size: {len(df)}")
print(f"DataFrame size after removing outliers: {len(df_no_outliers)}")
print(df_no_outliers.describe())
```

## Violin Plots:

Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.
Which product category tends to have the highest median price? Don't filter here by top categories.
Bar Charts:

Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
Which product category commands the highest average price? Don't filter here by top categories.

```python
# Identify the top 20 categories by count
top_categories = df_no_outliers['category'].value_counts().head(20).index

# Filter the DataFrame to include only these categories
df_top_categories = df_no_outliers[df_no_outliers['category'].isin(top_categories)]
df_top_categories
```

```python
plt.figure(figsize=(15, 8))
sns.violinplot(x='category', y='price', data=df_top_categories, palette='Pastel1', hue="category")
plt.title('Price Distribution Across Top 20 Product Categories')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Product Category')
plt.ylabel('Price')
plt.show()
```

```python
# Calculate median prices across all categories
category_medians = df_no_outliers.groupby('category')['price'].median()

# Identify the category with the highest median price
highest_median_category = category_medians.idxmax()
highest_median_price = category_medians.max()

print(f"The category with the highest median price is '{highest_median_category}' with a median price of {highest_median_price:.2f}.")
```

# Part 3: Investigating the Interplay Between Product Prices and Ratings
Objective: Analyze how product ratings (stars) correlate with product prices.

## Correlation Coefficients:

Calculate the correlation coefficient between price and stars.

Is there a significant correlation between product price and its rating?

```python
# Drop rows where 'price' or 'stars' might be missing
df_clean = df_no_outliers.dropna(subset=['price', 'stars'])
```

```python
correlation_matrix = df_clean[['price', 'stars']].corr()
price_stars_corr = correlation_matrix.loc['price', 'stars']

print(f"Correlation coefficient between price and stars: {price_stars_corr:.2f}")
```

```python
# Close to 1 or -1: Indicates a strong positive or negative correlation, respectively.
# Close to 0: Indicates little to no linear correlation.
```

## Visualizations:

Use a scatter plot to visualize the relationship between product rating and price. What patterns can you observe?

Use a correlation heatmap to visualize correlations between all numerical variables.

Examine if product prices typically follow a normal distribution using a QQ plot.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='stars', y='price', data=df_clean, alpha=0.7, color='skyblue')

# Add titles and labels
plt.title('Scatter Plot of Product Rating vs. Price')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df_clean.select_dtypes(include=np.number)

# Calculate the correlation matrix
corr = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
import scipy.stats as stats

# Generate a QQ plot
plt.figure(figsize=(8, 8))
stats.probplot(df_clean['price'], dist="norm", plot=plt)

# Add titles and labels
plt.title('QQ Plot for Product Prices')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.grid()
plt.show()

# Points closely following the theoretical line suggest that the prices are normally distributed.
# Deviation from the line indicates the presence of skewness or kurtosis (tails heavier/lighter than expected).
```
